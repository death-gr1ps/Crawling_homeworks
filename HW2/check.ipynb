```python
import json
import requests
import pandas as pd


def read_jsons(row_num, file_name):
    out = []
    with open(file_name, 'r', encoding='utf-8', errors='replace') as file:
        for i, line in enumerate(file):
            if i < row_num:
                out.append(json.loads(line))
            else:
                break
                
    return out
```

```python
json_res = read_jsons(20, "spider_hw2.jsonlines")
isbn_ls  = [json['isbn'] for json in json_res]
req_res  = list()

for isbn in isbn_ls:
    response = requests.get(f"http://localhost:8000/search_by_isbn?isbn={isbn}")
    req_res.append(response.json())
```
